```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## 侵蝕(Erosion)

```python
img = cv2.imread('b.jpg')
```

```python
kernel = np.ones((5,5), np.uint8)
erosin = cv2.erode(img, kernel, iterations=1)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(erosin)
```

```python
#迭代次數
plt.figure(figsize=(10,8))
for i in range(4):
    erosin = cv2.erode(img, kernel, iterations=i+1)
    plt.subplot(2,2,i+1)
    plt.title("iteration:"+ str(i+1))
    plt.imshow(erosin)
```

## 膨脹(Dilation)

```python
img = cv2.imread('a.jpg')
```

```python
kernel = np.ones((5,5), np.uint8)
dilation = cv2.dilate(img, kernel, iterations=2)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(dilation)
```

```python
#迭代次數
plt.figure(figsize=(10,8))
for i in range(4):
    dilation = cv2.dilate(img, kernel, iterations=i+1)
    plt.subplot(2,2,i+1)
    plt.title("iteration:"+ str(i+1))
    plt.imshow(dilation)
```

## 開運算及避運算(Opening & Closing)

```python
img = cv2.imread('c.jpg')
plt.imshow(img)
```

```python
kernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=1)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=1)

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.title('Original')
plt.imshow(img)
plt.subplot(2,2,3)
plt.title('Opening')
plt.imshow(opening)
plt.subplot(2,2,4)
plt.title('Closing')
plt.imshow(closing)
```

## 梯度運算

```python
img = cv2.imread('d.jpg')

kernel = np.ones((3,3), np.uint8)
dilation = cv2.dilate(img, kernel, iterations=3)
erosion = cv2.erode(img, kernel, iterations=3)

#膨脹 - 侵蝕
img_grad = dilation - erosion

plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_grad)
```

```python
#morphology
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel, iterations=2)
plt.imshow(gradient)
```

## 頂帽(Top Hat) = 原圖像 - 開運算圖像
## 黑帽(Black Hat) = 閉運算圖像 - 原圖像

```python
img = cv2.imread('c.jpg')
```

```python
kernel = np.ones((3,3), np.uint8)
#頂帽(Top Hat) = 原圖像 - 開運算圖像
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=3)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel, iterations=3)

plt.figure(figsize=(12,10))
plt.subplot(221)
plt.title('Original')
plt.imshow(img)
plt.subplot(223)
plt.title('opening')
plt.imshow(opening)
plt.subplot(224)
plt.title('TopHat')
plt.imshow(tophat)
```

```python
#黑帽(Black Hat) = 閉運算圖像 - 原圖像

closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=3)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel, iterations=3)

plt.figure(figsize=(12,10))
plt.subplot(221)
plt.title('Original')
plt.imshow(img)
plt.subplot(223)
plt.title('closing')
plt.imshow(closing)
plt.subplot(224)
plt.title('blackhat')
plt.imshow(blackhat)
```
